```python
import numpy as np
from sklearn.datasets import load_digits

digits = load_digits()
X, y = [np.array(x) for x in digits.data], digits.target
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.decomposition import PCA

from mapper.core import *
from mapper.cover import *
from mapper.clustering import *
from mapper.plot import *

lens = PCA(2).fit_transform(X)
#mapper_algo = MapperAlgorithm(cover=CubicCover(n=15, perc=0.25), clustering=KMeans(2, n_init='auto'))
mapper_algo = MapperAlgorithm(cover=CubicCover(n=10, perc=0.25), clustering=FirstSimpleGap())
mapper_graph = mapper_algo.fit_transform(X, lens)
```

```python
mapper_plot = MapperPlot(X, mapper_graph, iterations=50)
fig = mapper_plot.with_colors(colors=y, cmap='jet', agg=np.nanmedian).plot_interactive_2d(title='digit', width=512, height=512)
fig.show(config={'scrollZoom': True})
```

```python
from gtda.mapper.cover import CubicalCover
from sklearn.cluster import DBSCAN
from gtda.mapper.cluster import FirstSimpleGap

filter_func = PCA(2)
cover = CubicalCover(n_intervals=10, overlap_frac=0.5)
clusterer = FirstSimpleGap()

# Configure parallelism of clustering step
n_jobs = 1

# Initialise pipeline
pipe = make_mapper_pipeline(
    filter_func=filter_func,
    cover=cover,
    clusterer=clusterer,
    verbose=False,
    n_jobs=n_jobs,
)
fig = plot_static_mapper_graph(pipe, X)
```
